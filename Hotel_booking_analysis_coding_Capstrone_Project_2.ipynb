### Data Wrangling

#### Data wrangling code

Handling Null Values.

```python
Total = hotel_booking_data.isnull().sum().sort_values(ascending=False)
Percentage = (hotel_booking_data.isnull().sum()/hotel_booking_data.isnull().count()*100).sort_values(ascending=False)
Missingdata = pd.concat([Total,Percentage], axis=1, keys=['Total', 'Percentage'])
Missingdata.head(10)
```

From the output its clear that Company has a null percentage of 94%, and the agent as only 13.68%. We will drop the higher missing value column, so the company with 94% missing value will not be considered.

For Country the is 488 rows with null values comparing 488 to total numbers 0.5% so we will remove those rows only, alsochildren has four missing values we will fill with mean values, rest all are fine.

```python
hotel_booking_data = hotel_booking_data.drop(['company','agent'], axis=1)
hotel_booking_data = hotel_booking_data.dropna()
hotel_booking_data.info()
```

We have now 30 column with zero null values.

we will check other possiblilities here one of the threat that sum of all customer i.e Adult + children + babies should be greater than zero. lets see

total 170 rows are there having sum of adults, children and babies is zero will remove these rows

```python
hotel_booking_data = hotel_booking_data.drop(hotel_booking_data[(hotel_booking_data.adults+hotel_booking_data.babies+hotel_booking_data.children)==0].index)

#lets confirm
hotel_booking_data[(hotel_booking_data.adults+hotel_booking_data.babies+hotel_booking_data.children)==0].shape
```

Addition of required columns and processes.

```python
# month number column
month = {'January':1,'February':2, 'March':3, 'April':4, 'May':5, 'June':6, 'July':7, 'August':8, 'September':9, 'October':10, 'November':11, 'December':12}

hotel_booking_data['month_no']=0
for index,row in hotel_booking_data.iterrows():
  hotel_booking_data.loc[index,'month_no']=month[row['arrival_date_month']]
```

```python
# Adding arrival date
hotel_booking_data['dateInt']=hotel_booking_data['arrival_date_year'].astype(str) + hotel_booking_data['month_no'].astype(str).str.zfill(2)+ hotel_booking_data['arrival_date_day_of_month'].astype(str).str.zfill(2)
hotel_booking_data['Date'] = pd.to_datetime(hotel_booking_data['dateInt'], format='%Y%m%d')
```

```python
# Adding month year column
hotel_booking_data['month_year'] = pd.to_datetime(hotel_booking_data['Date']).dt.to_period('M')
```

```python
hotel_booking_data['total_guest']=(hotel_booking_data['adults']+hotel_booking_data['children']+hotel_booking_data['babies']).astype(int)
```

### Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables

#### Chart 1: Bar Plot

Distribution channel wise booking analysis.

```python
Distribution_chanel_analysis=hotel_booking_data.groupby('distribution_channel')['hotel'].count().reset_index().sort_values(by='hotel',ascending=False)
Distribution_chanel_analysis['percentage']=(Distribution_chanel_analysis['hotel']/(Distribution_chanel_analysis['hotel'].sum()))*100
Distribution_chanel_analysis.rename(columns={'hotel':'No. of booking'},inplace=True)    # Rename the column's name
Distribution_chanel_analysis
```

```python
Distribution_chanel_analysis.plot.bar(x='distribution_channel',y='percentage')
```

#### Chart 2: Line plot

Coumtry wise booking analysis

```python
country_wise=hotel_booking_data.groupby('country')['hotel'].count().reset_index().sort_values(by='hotel',ascending=False)
country_wise['percentage']=(country_wise['hotel']/country_wise['hotel'].sum())*100
country_wise.rename(columns={'hotel':'No. of booking'},inplace=True) # Rename columns
country_wise.head(10)  # Get top 10 countries
```

#### Chart 3: Pie Chart

Hotel Type wise booking analysis

```python
hotel_type_wise_booking_analysis=hotel_booking_data.groupby('hotel')['meal'].count().reset_index()
H_type = hotel_booking_data['hotel'].value_counts().tolist()
plt.pie(H_type,explode=(0,0.1,),labels=("City","Resort"),radius=2,shadow=True,autopct='%1.1f%%')
plt.show
```

#### Chart 4: Heatmap

Month wise booking analysis:

Through this analysis we will get the busiest month of the year.

```python
month_wise = hotel_booking_data.groupby(['month_year', 'hotel'])['meal'].count().unstack()

# Generating the heatmap
plt.figure(figsize=(10, 8))  # Adjust the figure size if needed
sns.heatmap(month_wise, cmap='YlGnBu', annot=True, fmt="d")
plt.title('Monthly Meal Count by Hotel')
plt.xlabel('Hotel')
plt.ylabel('Month-Year')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent overlapping labels
plt.show()
```

Cancellation analysis

```python
# Let's analyse the cancellation of booking for given dataset
Cancellation_analysis=hotel_booking_data.groupby('hotel')['is_canceled'].sum().reset_index()
Cancellation_analysis.rename(columns={'is_canceled':'No. of cancellation'},inplace=True)
Cancellation_analysis
```

```python
hotel_type_wise_booking_analysis
```

```python
Cancellation=pd.merge(Cancellation_analysis, hotel_type_wise_booking_analysis, on='hotel', how='inner')
Cancellation['percentage']=(Cancellation['No. of cancellation']/Cancellation['meal'])*100 # here we used meal column just for count purpose
Cancellation.rename(columns={'meal': 'Total no. of bookings'}, inplace=True)
Cancellation
```

#### Chart 8: Line chart

Conversion rate

```python
# Conversion rate
plt.figure(figsize=(20, 9))

# Filter data for City Hotel and Resort Hotel
city_hotel_conversion_data = hotel_booking_data[hotel_booking_data['hotel'] == 'City Hotel'].groupby('Date').size()
resort_hotel_conversion_data = hotel_booking_data[hotel_booking_data['hotel'] == 'Resort Hotel'].groupby('Date').size()

# Plotting line chart for City Hotel
plt.subplot(1, 2, 1)
city_hotel_conversion_data.plot(color='blue')  # Set color to blue
plt.title('Booking Conversion Rates over Time (City Hotel)')
plt.xlabel('Date')
plt.ylabel('Number of Bookings')

# Plotting line chart for Resort Hotel
plt.subplot(1, 2, 2)
resort_hotel_conversion_data.plot(color='orange')  # Set color to orange
plt.title('Booking Conversion Rates over Time (Resort Hotel)')
plt.xlabel('Date')
plt.ylabel('Number of Bookings')

plt.tight_layout()
plt.show()
```

#### Chart 9: Scatter Plot

Occupancy rate

```python
# Filter data for resort hotel and city hotel
resort_data = hotel_booking_data[hotel_booking_data['hotel'] == 'Resort Hotel']
city_data = hotel_booking_data[hotel_booking_data['hotel'] == 'City Hotel']

# Calculate occupancy rates for resort hotel and city hotel
resort_occupancy = (resort_data.groupby('Date')['is_canceled'].count() - resort_data.groupby('Date')['is_canceled'].sum()) / resort_data['reserved_room_type'].count() * 100
city_occupancy = (city_data.groupby('Date')['is_canceled'].count() - city_data.groupby('Date')['is_canceled'].sum()) / city_data['reserved_room_type'].count() * 100

# Plotting for Resort Hotel
plt.figure(figsize=(12, 6))
plt.scatter(resort_occupancy.index, resort_occupancy, label='Resort Hotel', color='blue', marker='o')
plt.title('Occupancy Rates over Time for Resort Hotel')
plt.xlabel('Date')
plt.ylabel('Occupancy Rate (%)')
plt.legend()
plt.grid(True)
plt.show()

# Plotting for City Hotel
plt.figure(figsize=(12, 6))
plt.scatter(city_occupancy.index, city_occupancy, label='City Hotel', color='red', marker='x')
plt.title('Occupancy Rates over Time for City Hotel')
plt.xlabel('Date')
plt.ylabel('Occupancy Rate (%)')
plt.legend()
plt.grid(True)
plt.show()
```
